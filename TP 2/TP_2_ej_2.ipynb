```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv(
    'C:\\Users\\nahue\\Desktop\\Agro 2\\Tp_2\\Ej_2.txt',
    delimiter = ' ',
    decimal = '.'
    )
```

```python
datos.head()
```

```python
datos['Etr'] = datos['ETP']*datos['KC']

agua_util_10 = 17
capacidad_campo_10 = agua_util_10/0.85
almacenaje_inicial_10 = agua_util_10*0.8
punto_marchitez_10 = capacidad_campo_10 - agua_util_10

agua_util = agua_util_10/100
capacidad_campo = capacidad_campo_10/100
almacenaje_inicial = almacenaje_inicial_10/100
punto_marchitez = punto_marchitez_10/100

agua_util_30 = agua_util*300
capacidad_campo_30 = capacidad_campo*300
almacenaje_inicial_30 = almacenaje_inicial*300
punto_marchitez_30 = punto_marchitez*300

print('En 10 cm: \n',
      capacidad_campo_10, '\n',
      agua_util_10, '\n',
      almacenaje_inicial_10, '\n',
      punto_marchitez_10, '\n',
      'En tita: \n',
      capacidad_campo, '\n',
      agua_util, '\n',
      almacenaje_inicial, '\n',
      punto_marchitez, '\n',
      'En 30 cm: \n',
      capacidad_campo_30, '\n',
      agua_util_30, '\n',
      almacenaje_inicial_30, '\n',
      punto_marchitez_30, '\n',
      )
```

```python
datos['variacion'] = datos['PP'] - datos['Etr']
datos['acumulada'] = datos['variacion'].cumsum()
datos.head(20)
```

```python
'''Asumo que las raíces medían 10cm hasta 3°noviembre y 20cm
hasta el 3° diciembre. Luego no crecen mas'''

raices = np.array([100, 100, 100,
          200, 200, 200,
          300, 300, 300,
          300, 300, 300,
          300, 300
         ])
datos['almacenaje sin riego'] = round(almacenaje_inicial*raices
                                      + datos['acumulada'],
                                      2
                                     )
datos.head(20)
```

```python
iterador = list(range(1, len(datos['almacenaje sin riego'])))
limite_riego = 0.5*agua_util*raices
datos['almacenaje con riego'] = datos['almacenaje sin riego']
datos['almacenaje con riego'][0] = 20
datos['riego'] = list(range(14))
datos['riego'][0] = 11.88
for i in iterador:
    datos['almacenaje con riego'][i] = (datos['almacenaje con riego'][i-1]
                                        + datos['variacion'][i]
                                       )
    datos['riego'][i] = 0
    if  datos['almacenaje con riego'][i] < limite_riego[i]: 
        datos['riego'][i] = (capacidad_campo*raices[i] 
                             - datos['almacenaje con riego'][i]
                            )
        datos['almacenaje con riego'][i] = capacidad_campo*raices[i]
    if  datos['almacenaje con riego'][i] > capacidad_campo*raices[i]:
        datos['almacenaje con riego'][i] = capacidad_campo*raices[i]
datos.head(15)
```
