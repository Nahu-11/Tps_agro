```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def e_s(t):
    e = 0.6108 * np.exp((17.27*t)/(t+237.3))
    e = e*10
    return e
def delta(es, t):
    delt = (4098*es) / (t+237.3)**2
    return delt
def gamma(P):
    gamma = (1930*P) / (2257000*0.622)
    return gamma
def EP(delta, Rn, es, e, u, gamma) :
    Ea = (es-e) * (0.27 + 0.233*u)
    Et = ((delta/(delta+gamma)) * Rn*0.9
          + (gamma/(delta+gamma)) * Ea
         )
    return Et
```

```python
datos['Presion'] = list(range(7))
datos['ET penman'] = list(range(7))
datos['Radiacion'] = datos['Rn']*86400/1000000
for i in datos['Presion']:
    datos['Presion'][i] = 1013
    e_s_p = e_s(datos['T'][i])
    delta_p = delta(e_s_p, datos['T'][i])
    gamma_p = gamma(datos['Presion'][i])
    datos['ET penman'][i] = EP(delta_p, datos['Radiacion'][i], e_s_p/10,
                               datos['e'][i]/10, datos['V'][i], gamma_p
                              )
datos['ETR'] = 1.25*datos['ET penman']
datos.head(10)
```

```python
datos['balance'] = list(range(7))
datos['balance'][0] = 950 - datos['ETR'][0]
iterador = list(range(1,7))
for i in iterador:
    datos['balance'][i] = (datos['balance'][i-1]
                           + datos['PP'][i]
                           - datos['ETR'][i]
                          )
datos.head(10)
```
