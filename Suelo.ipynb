```python
# Ejercicio 5
def onda(ts,a0,z,d,p,t0,t):
    return ts + a0 * np.exp(-z/d) * np.sin((2*3.14/p)*(t-t0)-z/d)
mod = np.array([0.01,0.5,0.9,1,1.1,1.5,2], dtype=float)
total = len(mod)
ts = 10 * mod
ao = 3 * mod
d = 40 * mod
t0 = 7
p = 24
z = 5
t = list(range(24))
resultado1 = list(range(24))
resultado2 = list(range(24))
resultado3 = list(range(24))
for x in range(total):
    for y in t:
        resultado1[y] = onda(ts[3],ao[x],z,d[3],p,t0,y)
        resultado2[y] = onda(ts[x],ao[3],z,d[3],p,t0,y)
        resultado3[y] = onda(ts[3],ao[3],z,d[x],p,t0,y)
    plt.figure(num = 1, figsize=(18,9))
    plt.plot(t, resultado1)
    plt.xlabel('Horas')
    plt.ylabel('Temperatura')
    plt.axis([0, 25, -5, 25])
    plt.title('Modifico amplitud')
    
    plt.figure(num = 2, figsize=(18,9))
    plt.plot(t, resultado2)
    plt.xlabel('Horas')
    plt.ylabel('Temperatura')
    plt.axis([0, 25, -5, 25])
    plt.title('Modifico temperatura promedio')
    
    plt.figure(num = 3, figsize=(18,9))
    plt.plot(t, resultado3)
    plt.xlabel('Horas')
    plt.ylabel('Temperatura')
    plt.axis([0, 25, -5, 25])
    plt.title('Modifico profundidad de amortiguacion')
```
